```python
#load libraries
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz,plot_tree
from IPython.display import Image
```

```python
df=pd.read_csv("/content/income_train.csv")
print(df)
```

```python
feature_cols=['age','education.num','hours.per.week']
feature=df[feature_cols] #features
Target=df.income #target variable
```

```python
#create decision tree classifer object
DT=DecisionTreeClassifier()
#train decision tree classifier
DT=DT.fit(feature,Target)
```

```python
#predict the response for test dataset
age=int(input("Enter age:"))
education_no=int(input("Enter education num:"))
hours_per_week=int(input("Enter hours_per_week:"))
pred=DT.predict([[age,education_no,hours_per_week]])
if pred==1:
  print("greater than 50,0000")
else:
  print("less than 50,000")
```

```python
plot_tree(DT) #plot_tree(trained model name)
```
